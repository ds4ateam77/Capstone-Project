## Hypothesis Testing

### Hypothesis 1

**Null hypothesis: There is a statistically significant difference between the value to cost ratio for players in different divisions.**

**Alternate hypothesis: There is no statistically significant difference between the value to cost ratio for players in different divisions.**

```python
endorsements_2019.head()
```

```python
endorsement_location = endorsements_2019[['Endorsment (millions)','Division']]
endorsement_location.head()
```

```python
#map location to numbers for hypothesis test
list_of_regions = endorsement_location['Division'].drop_duplicates().to_list()

print(list_of_regions)
list_of_regions_dict = {'Pacific':0, 'Atlantic':1, 'Northwest':2, 'Southwest':3, 'Central':4, 'Southeast':5}
print(list_of_regions_dict)
```

```python
endorsement_location['Division Number'] = endorsement_location['Division'].map(list_of_regions_dict)
endorsement_location.head()
```

```python
endorsement_location.groupby('Division').mean()\
.sort_values(by='Endorsment (millions)', ascending=False)['Endorsment (millions)'].plot(kind='bar')
```

Looking at the right skew of the data, it is apparent that there are two key players (Lebron James and Steph Curry) contributing the large discrpancy between Divisions. From now on, we will stay away from pure endorsements and look at value to cost ratio for each of the players.

```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
pacific = endorsements_2019[endorsement_location['Division'] == 'Pacific']['Value to cost']
atlantic = endorsements_2019[endorsement_location['Division'] == 'Atlantic']['Value to cost']
southwest = endorsements_2019[endorsement_location['Division'] == 'Southwest']['Value to cost']
central = endorsements_2019[endorsement_location['Division'] == 'Central']['Value to cost']
northwest = endorsements_2019[endorsement_location['Division'] == 'Northwest']['Value to cost']
southeast = endorsements_2019[endorsement_location['Division'] == 'Southeast']['Value to cost']
```

```python
type(pacific)
```

```python
f_oneway(pacific, atlantic, southwest, central, northwest, southeast)
```

From the Anova test, we can see that the p value is higher than 0.05 so we fail to reject the null hypothesis, we don't have enough evidence to reject it. Because the p value is higher than 0.05, there is no statistically significant difference between the means of the six independent Division groups. Note, in the event that we did have to reject the null hypothesis then, we would run a tukey test to see what the difference between the groups is.

```python
value_cost_location = endorsements_2019[['Value to cost','Division']]
```

```python
value_cost_location.groupby('Division').mean()\
.sort_values(by='Value to cost', ascending=False).plot(kind='bar')
```

### Hypothesis 2

**Null hypothesis: Players with shooting percentages in the top 50th percentile have the same value to cost ratios as players in the bottom 50th percentile.**

**Alternate hypothesis: Players with shooting percentages in the top 50th percentile have greater value to cost ratios than players in the bottom 50th percentile.**

Note: Use tail as either two-sided, one-sided, greater, or less in order to compare the two groups.

```python
endorsements_2019['Shooting_percentage'].describe()
```

```python
shooting_pct_median = endorsements_2019['Shooting_percentage'].median()
shooting_pct_median
```

```python
high_shooting_pct = endorsements_2019[endorsements_2019['Shooting_percentage'] > shooting_pct_median]
low_shooting_pct = endorsements_2019[endorsements_2019['Shooting_percentage'] < shooting_pct_median]
```

```python
low_shooting_pct.head(2)
```

```python
#Notes to self:

#use chisquared + contingency tables when we have two categorical variables
#hypothesis for chi squared would state whether or not the variables are dependent or independent

#use ttest when we have quantitative variables
#hypothesis would compare two populations 
#null: group 1 is the same as group 2

#anova test is used when you have more than two groups
#similar to ttest
#null: all of the groups are the same

#if the groups are not all the same...tukey hsd test
#similar to anova test
#looks at the differences between the groups if they are different
```

```python
import pingouin
pingouin.ttest(high_shooting_pct['Value to cost'], low_shooting_pct['Value to cost'], tail='greater')
```

The p-value is less than 0.05 so we must reject the null hypothesis and accept the alternate hypothesis. Therefore, players with the highest shooting percentage receive more endorsements than players with the lowerst shooting percentage.

### Hypothesis 4

**Null hypothesis: Players in the top 50th percentile in points per minute have the same value to cost ratio as players in the bottom 50th percentile in points per minute.**

**Alternate hypothesis: Players in the top 50th percentile in points per minute have a higher value to cost ratio than players in the bottom 50th percentile in points per minute.**

```python
ppg_2018_2019 = pd.read_csv('2018-19 Points Per Game Rankings.csv')
ppg_2018_2019.head()
```

```python
ppg_2018_2019['Player'] = ppg_2018_2019['Player'].str.split(pat= '\\', expand = True)[0]
ppg_2018_2019.head()
```

```python
ppg_2018_2019['PTS per MP'] = ppg_2018_2019['PTS▼']/ppg_2018_2019['MP']
ppg_2018_2019.head()
```

```python
ppm_2018_2019 = ppg_2018_2019[['Player', 'PTS per MP']]
ppm_2018_2019.head(2)
```

```python
endorsements_2019.head(2)
```

```python
endorsements_2019.head()
```

```python
ppm_endorsements_2019 = endorsements_2019.merge(ppm_2018_2019, on='Player')
ppm_endorsements_2019.head()
```

```python
ppm_median = ppm_endorsements_2019['PTS per MP'].median()
ppm_median
```

```python
high_ppm = ppm_endorsements_2019[ppm_endorsements_2019['PTS per MP'] > ppm_median]
low_ppm = ppm_endorsements_2019[ppm_endorsements_2019['PTS per MP'] < ppm_median]
```

```python
high_ppm.head(2)
```

```python
import pingouin
pingouin.ttest(high_ppm['Value to cost'], low_ppm['Value to cost'], tail='greater')
```

The p-value is higher than 0.05 so we cannot reject the null hypothesis. Therefore, there isn't significant statistical evidence to reject that players in the top 50th percentile in points per minute have the same value to cost ratio as players in the bottom 50th percentile in points per minute.
